```python
# Complete the function below.
#time started 11:45 2/18 ended 12:01 while copying some code above
def find_all_arrangements(n):
    def get_updated_covered_spaces(old_covered_spaces, r, c):
        covered_spaces = set()
        #fill rows
        for col in range(n):
            covered_spaces.add((r, col))

        #fill cols
        for row in range(n):
            covered_spaces.add((row, c))

        #fill diagonal bottom left to top right
        row = r
        col = c
        while row >= 0 and col < n:
            covered_spaces.add((row, col))
            row -= 1
            col += 1
        row = r
        col = c
        while row < n and col >= 0:
            covered_spaces.add((row, col))
            row += 1
            col -= 1
        #fill diagonal top left to bottom right
        row = r
        col = c
        while row < n and col < n:
            covered_spaces.add((row, col))
            row += 1
            col += 1
        row = r
        col = c
        while row >= n and col >= 0:
            covered_spaces.add((row, col))
            row -= 1
            col -= 1

        return covered_spaces.union(old_covered_spaces)

    def helper(board, row, covered_spaces, collector):
        if row == n:
            transformed_board = []
            for target_col in board:
                row = []
                for col in range(n):
                    if col == target_col:
                        row.append('q')
                    else:
                        row.append('-')
                transformed_board.append("".join(row))
            collector.append(transformed_board)
        else:
            for col in range(n):
                if (row,col) not in covered_spaces:
                    board.append(col)
                    helper(board, 
                           row + 1, 
                           get_updated_covered_spaces(covered_spaces, row, col),
                          collector)
                    board.pop()
    coll = []
    helper([], 0, set(), coll)
    
    return coll
        
            
    
```
