```python
# Complete the function below.
#start time: 2:18
def generate_palindromic_decompositions(s):
    
    def isPalindrome(sub_string):
        start = 0
        end = len(sub_string) - 1
        
        while start < end:
            if sub_string[start] != sub_string[end]:
                return False
            start += 1
            end -= 1
        return True
    
    #idx = current character
    #soFar=curent output
    #curr_substr=current substring we are analyzing
    coll = []
    def helper(s, soFar, curr_substr, idx):
        print("soFar:", soFar)
        print("cur_substr:", curr_substr)
        if idx == len(s):
            if len(curr_substr) > 0 and isPalindrome(curr_substr):
                soFar.append("".join(curr_substr))
                coll.append("|".join(soFar))
                soFar.pop()
        else:
            new_substr = curr_substr + [s[idx]]
            #check if substring is a palindrome
            if isPalindrome(new_substr):
                soFar.append("".join(new_substr))
                helper(s,soFar, [], idx + 1)
                soFar.pop()
            #add to substring
            helper(s, soFar, new_substr, idx + 1)
            
    helper(s, [], [], 0)
    
    return coll
```

```python
generate_palindromic_decompositions("abracadabra")
```
